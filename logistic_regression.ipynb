# import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Add the CSV file

```python
df=pd.read_csv('insurance_data.csv')
df.head ()
```

# Graphe show the dataset

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
```

# Split the data into trainning and Testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['age']], df.bought_insurance, test_size=0.8)
# Correctly assign the returned values from train_test_split
```

```python
print(X_train)
```

# Select the model that will use

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

#predict the values by adding the X_test

```python
y_pred=model.predict(X_test)
```

```python
print(y_pred)
```

```python
model.predict_proba(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import math   # to make the rule of sigmoid
```

```python
def sigmoid(X):
  sig=1/(1+math.exp(-X))
  return sig
```

```python
def predict_function(age):
  # z=m*age+b        #Z= 0.32 *age -12.80
  z=0.32*age-12.80
  y_pred=sigmoid(z)
  return y_pred
```

# we will give the model age to predict if it can buy insurance or not

```python
age=35
y_pred=predict_function(age)
print(y_pred)
```


# ** 0.167 is less than 0.5 which means person with 35 age will not buy insurance**

```python
age =60
y_pred=predict_function(age)
print(y_pred)
```
